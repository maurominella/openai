# Simple "Responses API" call using the `openai` package (Azure OpenAI)

This example shows how to:
- Load environment variables  
- Initialize the `AzureOpenAI` client  
- Send a minimal `responses.create()` request  
- Print the generated text output  

The goal is to demonstrate the lowest‑level usage of the Responses API before moving on to conversations, threads, and agent-style interactions.

# Constants and Libraries

```python
import os
from IPython.display import Markdown, display
from dotenv import load_dotenv # package python-dotenv
from openai import AzureOpenAI # package openai

if not load_dotenv("./../../config/credentials_my.env"):
    print("Environment variables not loaded, cell execution stopped")

else:
    print("Environment variables have been loaded ;-)")

messages_system = "You are an AI assistant that helps people find information"
messages_user = "Tell me five good names for my new pizzeria"
```

# Create a client to connecto Azure OpenAI service and deployment

```python
client = AzureOpenAI(
  api_version = os.getenv("AZURE_OPENAI_API_VERSION")
  # api_key = os.getenv("AZURE_OPENAI_API_KEY"),
  # azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
)
print(f"OpenAI Endpoint base_url: {client.base_url}")
```

# Invoke the client and print the response

```python
response = client.responses.create(
    model=os.environ["MODEL_DEPLOYMENT_NAME"],
    input=messages_user
)
print(response.output[0].content[0].text)
```
