# [Autogen Conversation Patterns](https://microsoft.github.io/autogen/0.2/docs/tutorial/conversation-patterns/)
Four conversation pattern in Autogen:
1. **Two-agent chat**: the simples form of conversation pattern hwhere two agents chat with each other using Autogen
2. **Sequential chat**: a series of conversations between two agents, linked by a carryover mechanisms that transfers the summary of the preio chat to the context of the subsequent chat
3. **Group chat**: a broader conversation that includes more than two agents, which is beneficial for tasks necessitating collaboration among agents.
4. **Nested Chat**: package a workflow into a single agent for reuse in a larger workflow.

# Sequential Chats
The name of this pattern is self-explanatory – it is a sequence of chats between two agents, chained together by a mechanism called carryover, which brings the summary of the previous chat to the context of the next chat.<br/><br/>
This pattern is useful for complex task that can be broken down into interdependent sub-tasks. The figure below illustrate how this pattern works:<br/><br/>


````python
# The Report Agent is responsible for creating a report by extracting insights from the chat history
report_agent = autogen.ConversableAgent (
    name = "Report_Agent",
    system_message = "You are responsible for creating a report by extracting insights from the chat history.",
    llm_config=llm_config,
    human_input_mode="NEVER"
)

# The Researcher Agent discovers the potential of the topic.
researcher_agent = autogen.ConversableAgent (
    name = "Researcher_Agent",
    system_message = "You explore and describe the potential capabilities and advancements of the topic.",
    llm_config=llm_config,
    human_input_mode="NEVER"
)

# The Ethicist Agent evaluates the ethical implications of the topic.
ethicist_agent = autogen.ConversableAgent (
    name = "Ethicist_Agent",
    system_message = "You evaluate the ethical implications of the topic, based on the research findings.",
    llm_config=llm_config,
    human_input_mode="NEVER"
)

# The Economist Agent analyzes the economic impact of the topic.
economist_agent = autogen.ConversableAgent (
    name = "Economist_Agent",
    system_message = "You analyze the economic impact of the topic, based on the ethical evaluations.",
    llm_config=llm_config,
    human_input_mode="NEVER"
)

# The Policy Maker Agent develops policies based on the finding of the previous agents.
policy_maker_agent = autogen.ConversableAgent (
    name = "Policy_Maker_Agent",
    system_message = "You develop policies to manage the topic, based on the economic analysis.",
    llm_config=llm_config,
    human_input_mode="NEVER"
)
````

````python
%%time

# Here we start a sequence of two-agent chats.
# Each element in the list is a dictionary that specifies the argument
# for the initiate_chat method.

sequential_chat_results = report_agent.initiate_chats(
    [
        {
            "recipient": researcher_agent,
            "message": topic,
            "max_turns": 1,
            "summary_method": "last_msg"
        },
        {
            "recipient": ethicist_agent,
            "message": "Evaluate the ethical implications of the topic",
            "max_turns": 1,
            "summary_method": "last_msg"
        },
        {
            "recipient": economist_agent,
            "message": "Analyze the economic impact of the topic",
            "max_turns": 1,
            "summary_method": "last_msg"
        },
        {
            "recipient": policy_maker_agent,
            "message": "Develop policies to manage the topic",
            "max_turns": 1,
            "summary_method": "last_msg"
        }
    ]
)
````

````python
for scr in sequential_chat_results:
    display(Markdown(scr.summary))
````

# Tool Calls

````python
def elevation_to_power(base: float, power: float) -> dict:
    """ Calculates the result of <base> elevated to <power> """
    import json
    elevation_to_power_result = {
        "result": float(base) ** float(power)
    }
    return json.dumps(elevation_to_power_result)


def current_date() -> dict:
    """ Returns the current date """
    import datetime
    current_date = {
        "current_date": datetime.date.today().strftime("%Y-%m-%d")
    }
    return json.dumps(current_date)


def get_days_difference(date_1:str, date_2:str) -> dict:
    """ calculate the distance (in days) between two dates """
    import json, random
    from dateutil.parser import parse
    
    days = {
        "days": abs((parse(date_2) - parse(date_1)).days) 
    }
    return json.dumps(days)


def web_search(question:str) -> dict:
    """ Does a WEB search and returns the content of first page found """
    from langchain.utilities import BingSearchAPIWrapper
    bing_search = BingSearchAPIWrapper()
    search_result = {
        "search_result": bing_search.run(question)
    }
    return json.dumps(search_result)

print(elevation_to_power(72.5,-3.5))
print(current_date())
print(get_days_difference("2025-12-25","2024-12-25"))
print(web_search("quali sono i film in programmazione per stasera?"))
````

````python
from duckduckgo_search import DDGS
from typing import Annotated
import datetime

def save_report (reort: str) -> str:
    try:
        file_name = f'report_{str(datetime.datetime.now().strftime("%Y%m%d_%H:%M:%S"))}.md'
        with open(file_name, "w") as f:
            f.write(report)
        return "Report saved successfully"
    except Exception as e:
        return "Failed to save the report due to the following error: {e}"
````

````python
def browse_web (query: str) -> str:
    try:
        with DDGS() as ddgs:
            results = [r for r in ddgs.text(query, max_results=1)]
            return results if results else "Not Found"
    except Exception as e:
        print (e)
````

````python
def web_search_tool (query: Annotated[str, 'Query string containing information that you want to search using the Internet']) -> str:
    result = browse_web(query)
    return result
````

````python
def save_report_tool (report: Annotated[str, 'Report in markdown format']) -> str:
    result = save_report(report)
    return result
````
